```python
from elasticsearch import Elasticsearch
es = Elasticsearch([{'host':'210.107.192.201', 'port':9200}])
#http://210.107.192.201/
#http://202.30.23.40/

import csv
import os
import pandas as pd

qryexps = open('2015_mix_topic_new.csv', 'rb')     
qryexps_read = qryexps.readlines()
qryexps_list = list(qryexps_read)
qryexps.close()  
#print(qryexps_list[0])


folder_name = "DFR_2015_Weight_1000"
index_name = "DFR summary"
#index_name2 = "dfr_2014_topic{0}".format(topicNum)
index_name2 = "dfr_total_weight"
#analyzer_name = "my_{0}_analyzer".format(index_name)
#types = "description"
file_name = "smry2015"
types = "summary"
newpath = './{0}/'.format(folder_name)
if not os.path.exists(newpath): os.makedirs(newpath)
title = "lamdarun03.txt"
target = open(newpath+title, 'w')

def rank_drf_2015(topicNum):
    

    df = pd.read_csv('{0}.csv'.format(file_name))

    if df.iloc[topicNum-1, 1]:
        print df.iloc[topicNum-1, 1]
        query = df.iloc[topicNum-1, 1]
    else:
        print("no summary")
    #print(title)
    label = "topicNum, rank, pmcid, score"
    print(label)
    
    #target.write(label + '\n')
    qry_expansion = qryexps_list[topicNum-1]
    query = query + " " + qry_expansion
    query = query.lower()
    query = query.replace(r"/",',')
    query = query.replace(r"[",' ')
    query = query.replace(r"]",' ')
    print(query)
    res = es.search(index=index_name2,doc_type="article", q=query ,size=1000, request_timeout=100000)
    count = 0
    for doc in res['hits']['hits']:
        count += 1
        #print("{0:.2f}".format(a))
        score = doc['_score']
        score = "{0:.4f}".format(score)
        text = '{0} Q0 {1} {2} {3} lamdarun03'.format(topicNum, doc['_source']['pmcid'], count, score)
        print (text)
        target.write(text + '\n')
    
for i in range(1, 31):
    rank_drf_2015(i)

target.close()
```
