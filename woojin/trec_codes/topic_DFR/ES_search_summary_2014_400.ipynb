```python
from elasticsearch import Elasticsearch
es = Elasticsearch([{'host':'210.107.192.201', 'port':9200}])
#http://210.107.192.201/
#http://202.30.23.40/

import csv
import os
import pandas as pd

qryexps = open('2014_mix_topic_new.csv', 'rb')     
qryexps_read = qryexps.readlines()
qryexps_list = list(qryexps_read)
qryexps.close()  
#print(qryexps_list[0])

def rank_drf_2015(topicNum):
    folder_name = "DFR_2014_topic_400_Sample2"
    index_name = "DFR summary"
    #index_name2 = "dfr_2014_topic{0}".format(topicNum)
    #index_name2 = "dfr_edge_2014ans_topic{0}".format(topicNum)
    index_name2 = "dfr_2014_sample"
    #analyzer_name = "my_{0}_analyzer".format(index_name)
    #types = "description"
    file_name = "summary"
    types = "summary"

    df = pd.read_csv('{0}.csv'.format(file_name))

    if df.iloc[topicNum-1, 1]:
        print df.iloc[topicNum-1, 1]
        query = df.iloc[topicNum-1, 1]
    else:
        print("no summary")

    newpath = './{0}/'.format(folder_name)
    if not os.path.exists(newpath): os.makedirs(newpath)
    title = "{0}{1}.csv".format(index_name, topicNum)
    #print(title)
    label = "topicNum, rank, pmcid, score"
    print(label)
    target = open(newpath+title, 'w')
    target.write(label + '\n')
    qry_expansion = qryexps_list[topicNum-1]
    query = query + " " + qry_expansion
    query = query.lower()
    query = query.replace(r"/",',')
    query = query.replace(r"[",' ')
    query = query.replace(r"]",' ')
    print(query)
    res = es.search(index=index_name2,doc_type="article", q=query ,size=400, request_timeout=100000)
    count = 0
    for doc in res['hits']['hits']:
        count += 1
        text = '{0}, {1}, {2}, {3}'.format(topicNum, count, doc['_source']['pmcid'], doc['_score'] )
        print (text)
        target.write(text + '\n')
    target.close()

    
for i in range(1, 31):
    rank_drf_2015(i)
```
